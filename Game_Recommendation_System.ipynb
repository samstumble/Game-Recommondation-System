```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler
```

```python
games = pd.DataFrame({
    "game": [
        "GTA V", "Cyberpunk 2077", "Red Dead Redemption 2",
        "Valorant", "Call of Duty Warzone",
        "The Witcher 3", "Forza Horizon 5", "Minecraft"
    ],
    "genre": [
        "Action", "RPG", "Action",
        "Shooter", "Shooter",
        "RPG", "Racing", "Sandbox"
    ],
    "platform": [
        "PC", "PC", "PC",
        "PC", "PC",
        "PC", "PC", "PC"
    ],
    "single_player": [1, 1, 1, 0, 0, 1, 1, 1],
    "multiplayer": [1, 0, 0, 1, 1, 0, 1, 1],
    "graphics_quality": [5, 5, 5, 3, 4, 5, 5, 2],
    "avg_playtime": [60, 40, 70, 120, 90, 80, 50, 200]
})
```

```python
games_encoded = pd.get_dummies(games, columns=["genre", "platform"])
```

```python
scaler = MinMaxScaler()
games_encoded[["graphics_quality", "avg_playtime"]] = scaler.fit_transform(
    games_encoded[["graphics_quality", "avg_playtime"]]
)
```

```python
user_profile = {
    "single_player": 1,
    "multiplayer": 0,
    "graphics_quality": 5,
    "avg_playtime": 70,
    "genre_RPG": 1,
    "genre_Action": 1,
    "genre_Shooter": 0,
    "genre_Racing": 0,
    "genre_Sandbox": 0,
    "platform_PC": 1
}
user_df = pd.DataFrame([user_profile])

# Align columns with game dataset
user_df = user_df.reindex(columns=games_encoded.columns, fill_value=0)
```

```python
similarity_scores = cosine_similarity(
    games_encoded.drop("game", axis=1),
    user_df.drop("game", axis=1, errors="ignore")
)

games["similarity_score"] = similarity_scores
```

```python
recommendations = games.sort_values(
    by="similarity_score",
    ascending=False
).head(5)

recommendations
```
